Consider the gridworld environment shown below (with almost everything the same as the previous homework): the agent can move in one of four directions {Up, Down Left, Right}. The agent has reward of -1 each time it moves, gets a reward +10 from reaching the goal, and a reward -10 from falling in the trap.
Entering either the goal or the trap state ends the episode. The agent stays in place when hitting the walls or the block in the middle of the grid. The discount factor is set to delta = 1.

Differences from the Gridworld in HW1: To be consistent with our unifying framework of episodic and continuing tasks, we now have a terminal state s_inf (the triangle state): the agent transitions to it after entering either the goal or trap states. Note also how the rewards +10 are collected while
transitioning to s_inf. Also, this time the agent is reliable.
